## 圖像 的 基本操作
### 1.畫 一個 物體


- `cv2.line(image, startPoint, endPoint, rgb, thinkness)`
- `cv2.rectangle(image, topLeft, bottomRight, rgb, thinkness)`
- `cv2.circle(image, center, radius, rgb, thinkness)`
- `cv2.ellipse(image, center, axes, angle, startAngle, endAngle, rgb, thinkness)`

```python
# 創建 一個 黑色圖像
# Create a black image
import numpy as np
img2 = np.zeros((512,512,3), np.uint8)
plt.imshow(img2)
```

```python
# cv2.line 繪製 一條線（image，startPoint，endPoint，rgb，thickness）
# Draw a line using cv2.line(image, startPoint, endPoint, rgb, thinkness)

cv2.line(img2, (0,0), (511,511), (255,0,0), 5)

# => 對角紅線，厚度為 5 像素
# => Diagonal red line with thickness of 5 px

# cv2.rectangle（image，topLeft，bottomRight，rgb，thickness）繪製 一個矩形
# Draw a rectangle using cv2.rectangle(image, topLeft, bottomRight, rgb, thinkness)

cv2.rectangle(img2, (384,0), (510,128), (0,255,0), 3)

# => 綠色矩形，厚度為 3 像素
# => Green rectangle with thickness of 3 px

# cv2.circle 繪製一個圓（圖像，中心，半徑，rgb，厚度）
# Draw a circle using cv2.circle(image, center, radius, rgb, thinkness)

cv2.circle(img2, (447,63), 63, (0,0,255), -1)

# ＃=>藍色圓圈 ( 厚度為 -1)
# => Blue filled circle(note that the thinkness is -1)


# cv2.ellipse 繪製 一個橢圓（圖像，中心，軸，角度，startAngle，endAngle，rgb，厚度）
# Draw a ellipse using cv2.ellipse(image, center, axes, angle, startAngle, endAngle, rgb, thinkness)

cv2.ellipse(img2, (256,256), (100,50), -45, 0, 180, (255,0,0), -1)

# => 紅色寬 下半橢圓
# => Red wide down-half ellipse

plt.imshow(img2)
```

```python
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);
# 顯示找到的 面數
# print the number of faces found

print('Faces found: ', len(faces))
```

```python
#go over list of faces and draw them as rectangles on original colored img
# 在面部 列表上，並將它們繪製為 原始彩色圖像 上的 矩形 

for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
# conver image to RGB and show image
# 將圖像轉換為RGB並顯示圖像

plt.imshow(cv2.cvtColor(test1,cv2.COLOR_BGR2RGB))
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1):
    # 製作一份傳遞的圖像副本，這樣就不會改變傳遞的圖像
    # just making a copy of image passed, so that passed image is not changed
    
    
    img_copy = colored_img.copy()
    
    # 將測試圖像轉換為灰度圖像，因為opencv人臉檢測器需要灰度圖像
    # convert the test image to gray image as opencv face detector expects gray images
    
    
    gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)
    
    # 讓我們檢測多尺度（一些圖像可能比其他圖像更接近相機）圖像
    #let's detect multiscale (some images may be closer to camera than others) images
    
    
    faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5);
    
    # 遍歷面部列表並將它們繪製為原始彩色圖像上的矩形
    #go over list of faces and draw them as rectangles on original colored img
    
    
    for (x, y, w, h) in faces:
        cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 5)
        
    return img_copy
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#加載另一張圖片 
#load another image

test2 = cv2.imread('data/test_cv06.jpg')

#  call 檢測 面部 的功能
#  call our function to detect faces

faces_detected_img = detect_faces(haar_face_cascade, test2)
print('Faces found: ', len(faces_detected_img))


#  call 檢測 面部 的功能
# conver image to RGB 並 顯示圖像

plt.imshow(convertToRGB(faces_detected_img))
```

```python
# 載入 用於 lbpcascade 的 cascade classifie 級聯分類器 訓練文件
# load cascade classifier training file for lbpcascade

lbp_face_cascade = cv2.CascadeClassifier('data/lbpcascade_frontalface.xml')


#載入 測試圖像
#load test image

test2 = cv2.imread('data/test_cv06.jpg')

# call 檢測 面部 的功能
# call our function to detect faces

faces_detected_img = detect_faces(lbp_face_cascade, test2)

print('Faces found: ', len(faces_detected_img))

#conver image to RGB 並 顯示圖像
#conver image to RGB and show image

plt.imshow(convertToRGB(faces_detected_img))
```
